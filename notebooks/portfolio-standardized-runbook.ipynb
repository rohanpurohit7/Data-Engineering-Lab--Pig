## Step 1 — Environment Setup

The original lab shows Pig installed under `/opt`, `.bashrc` updated with Pig and Hadoop paths, and Hadoop configuration paths added. This establishes the runtime context before the scripts execute.


```python
# Representative historical setup from the lab
cd /opt
cd pig-0.16.0
export PIG_HOME=/opt/pig-0.16.0
export HADOOP_CONF_DIR=/opt/hadoop-2.7.3/etc/hadoop
export PIG_CLASSPATH=/opt/hadoop-2.7.3/etc/hadoop
```

## Step 2 — Start Hadoop Services

The runbook starts HDFS and YARN, then validates the Java processes using `jps`. The process list is evidence that NameNode, DataNode, ResourceManager, NodeManager, and JobHistoryServer are available for distributed execution.


```python
start-dfs.sh
start-yarn.sh
mr-jobhistory-daemon.sh start historyserver
jps
```

## Step 3 — Script 1: Query Phrase Popularity

Script 1 processes a search-query log and finds phrases that occur with high frequency during certain times of day. This is a batch feature-extraction pattern: parse records, group by phrase/time feature, aggregate counts, and write ranked output.


```python
pig script1-hadoop.pig
hadoop fs -ls /user/hadoop/script1-hadoop-results
```

## Step 4 — Script 2: Temporal Query Phrase Popularity

Script 2 compares phrase frequency across two time periods separated by twelve hours. The analytical theme is temporal drift: identifying query behavior that changes across time windows.


```python
pig script2-hadoop.pig
hadoop fs -ls /user/hadoop/script2-hadoop-results
hadoop fs -cat /user/hadoop/script2-hadoop-results/part-r-00000 | head
```
